```python
#10. How can you have multiple constructors in a Python class? Explain with an example.
#ans:
'''Python does not support explicit multiple constructors, yet there are some ways using which multiple constructors can be achieved.
If multiple __init__ methods are written for the same class, then the latest one overwrites all the previous constructors and the reason for this can be,
python stores all the function names in a class as key in a dictionary so, when a new function is defined with the same name, the key remains the same 
but the value gets overridden by the new function body.'''
#example:
class actors:
    def __init__(self):
        print("a")

    def __init__(self):
        print(":B")
    def __init__(self):
        print("C")
girl=actors()
"it prints only latest one that is C "
#Need for multiple constructors
'''Multiple constructors are required when one has to perform different actions on the instantiation of a class. 
This is useful when the class has to perform different actions on different parameters. The class constructors 
can be made to exhibit polymorphism in three ways which are listed below.

1.Overloading constructors based on arguments.
2.Calling methods from __init__.
3.Using @classmethod decorator.
This article explains how to have multiple constructors in a clean and Pythonic way with examples.'''
'''Overloading constructors based on arguments
The constructor overloading is done by checking conditions for the arguments passed and performing required actions.
For example, consider passing an argument to the class sample, 

*If the parameter is an int, the square of the number should be the answer.
*If the parameter is a String, the answer should be “Hello!!”+string.
*If the parameter is of length greater than 1, the sum of arguments should be stored as the answer.'''
#example
class eval_equations:

	# basic constructor
	def __init__(self, a):
		self.ans = a

	# expression 1
	@classmethod
	def eq1(cls, args):
	
	# create an object for the class to return
		x = cls((args[0]*args[0])+(args[1]*args[1])-args[2])
		return x

	# expression 2
	@classmethod
	def eq2(cls, args):
		y = cls((args[0]*args[0])-(args[1]*args[1]))
		return y

	# expression 3
	@classmethod
	def eq3(cls, args):
		temp = 0

		# square of each element
		for i in range(0, len(args)):
			temp += args[i]*args[i]

		temp = temp/max(args)
		z = cls(temp)
		return z


li = [[1, 2], [1, 2, 3], [1, 2, 3, 4, 5]]
i = 0

# loop to get input three times
while i < 3:

	inp = li[i]

	# no.of.arguments = 2
	if len(inp) == 2:
		p = eval_equations.eq2(inp)
		print("equation 2 :", p.ans)

	# no.of.arguments = 3
	elif len(inp) == 3:
		p = eval_equations.eq1(inp)
		print("equation 1 :", p.ans)

	# More than three arguments
	else:
		p = eval_equations.eq3(inp)
		print("equation 3 :", p.ans)

	#increment loop	 
	i += 1
```

```python
# 19. Explain the use of constructor chaining in Python. Provide a practical example.
#ans:
'''Constructors are used for instantiating an object. The task of the constructor is to assign value
to data members when an object of the class is created.

Constructor chaining is the process of calling one constructor from another constructor.
Constructor chaining is useful when you want to invoke multiple constructors,
one after another, by initializing only one instance.

In Python, constructor chaining is convenient when we are dealing with inheritance. When an instance
of a child class is initialized, the constructors of all the parent 
classes are first invoked and then, in the end, the constructor of the child class is invoked.

Using the super() method we can invoke the parent class constructor from a child class.'''
class Vehicle:
    # Constructor of Vehicle
    def __init__(self, engine):
        print('Inside Vehicle Constructor')
        self.engine = engine

class Car(Vehicle):
    # Constructor of Car
    def __init__(self, engine, max_speed):
        super().__init__(engine)
        print('Inside Car Constructor')
        self.max_speed = max_speed

class Electric_Car(Car):
    # Constructor of Electric Car
    def __init__(self, engine, max_speed, km_range):
        super().__init__(engine, max_speed)
        print('Inside Electric Car Constructor')
        self.km_range = km_range

# Object of electric car
ev = Electric_Car('1500cc', 240, 750)
print(f'Engine={ev.engine}, Max Speed={ev.max_speed}, Km range={ev.km_range}')
 
```

```python
#12. Explain the concept of property decorators in Python and how they relate to encapsulation.
#ans:
'''In Python, property decorators are a way to implement getter and setter methods for class attributes in a more concise and elegant manner.
They allow you to define special methods that are called when you access or modify the value of an attribute. Property decorators are used to
create properties, which are special attributes that have getter, setter, and deleter methods associated with them.

Here's a brief explanation of the property decorator and how it relates to encapsulation:

Getter Method:
The @property decorator is used to define a getter method. It allows you to access the value of an attribute as if it were a regular attribute,
without calling a method explicitly.

Setter Method:
The @<attribute>.setter decorator is used to define a setter method. It is called when you try to modify the value of an attribute. This allows
you to perform additional checks or operations before setting the value.

Deleter Method:
The @<attribute>.deleter decorator is used to define a deleter method. It is called when you use the del statement to delete an attribute. This
allows you to perform cleanup operations before removing the attribute.

Now, let's see an example to illustrate the use of property decorators for encapsulation:'''
class Student:
    def __init__(self, name, age):
        self._name = name  # Convention: Single underscore for "protected" attribute
        self._age = age

    @property
    def name(self):
        """Getter method for the 'name' attribute."""
        return self._name

    @name.setter
    def name(self, new_name):
        """Setter method for the 'name' attribute."""
        if isinstance(new_name, str):
            self._name = new_name
        else:
            print("Invalid name. Please provide a string.")

    @property
    def age(self):
        """Getter method for the 'age' attribute."""
        return self._age

    @age.setter
    def age(self, new_age):
        """Setter method for the 'age' attribute."""
        if isinstance(new_age, int) and new_age >= 0:
            self._age = new_age
        else:
            print("Invalid age. Please provide a non-negative integer.")

# Example Usage:
if __name__ == "__main__":
    student = Student(name="Alice", age=20)

    # Using the getter methods
    print(f"Name: {student.name}, Age: {student.age}")

    # Using the setter methods
    student.name = "Bob"
    student.age = 25

    print(f"Updated Name: {student.name}, Updated Age: {student.age}")
```

```python
'''Data hiding is a concept in object-oriented programming that refers to the practice of restricting access to the internal details of an object and
exposing only a controlled interface to the outside world. It is a key aspect of encapsulation, one of the fundamental principles of OOP.

In data hiding, the internal state of an object (its attributes or properties) is made private, and access to these attributes is restricted to 
specific methods within the class. This ensures that the object's internal representation is not directly accessible from external code. Instead, 
external code interacts with the object through a well-defined set of public methods, providing a higher level of abstraction and allowing the class to manage its internal state in a controlled manner.

Here are a few reasons why data hiding is important in encapsulation:
=======================================================================
Encapsulation of Implementation Details:
Data hiding allows the encapsulation of the implementation details of a class. By making attributes private, the internal structure of an object is 
hidden from the outside world. This helps in achieving information hiding and reduces the complexity of the code that interacts with the object.

Controlled Access:
==================
By restricting direct access to the internal attributes, data hiding enables controlled access through methods. This means that external code must
go through designated methods to interact with the object's state. This control allows the class to enforce rules, validations, and constraints on 
how the data is accessed and modified.

Security and Integrity:
===========================
Hiding internal data helps in enhancing security by preventing unauthorized access and modification. It reduces the risk of unintentional data 
corruption or manipulation by external code, ensuring the integrity of the object's state.

Facilitates Maintenance and Refactoring:
=========================================
Encapsulation with data hiding makes it easier to maintain and refactor code. Since the internal details are hidden, changes to the implementation
can be made without affecting the external code that relies on the class. This promotes modularity and simplifies the process of updating or
extending the system.

Now, let's illustrate data hiding with an example:'''
#ans:
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute
        self.__balance = initial_balance        # Private attribute

    def deposit(self, amount):
        """Deposit money into the account."""
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid deposit amount. Please enter a positive value.")

    def withdraw(self, amount):
        """Withdraw money from the account."""
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        elif amount <= 0:
            print("Invalid withdrawal amount. Please enter a positive value.")
        else:
            print("Insufficient funds. Withdrawal not allowed.")

    def get_balance(self):
        """Get the current account balance."""
        return self.__balance

    def get_account_number(self):
        """Get the account number."""
        return self.__account_number

# Example Usage:
if __name__ == "__main__":
    account = BankAccount(account_number="123456789", initial_balance=1000)

    # Attempting to access private attributes directly
    # This would raise an AttributeError
    # Uncomment the following lines to see the error
    # print(account.__account_number)
    # print(account.__balance)

    # Accessing attributes through public methods
    print(f"Account Balance: ${account.get_balance()}")
    print(f"Account Number: {account.get_account_number()}")
```

```python
'''4. Create a Python class hierarchy for a banking system with various account types (e.g., savings, checking,
credit card) and demonstrate polymorphism by implementing a common `withdraw()` method.'''
#ans:
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawal of ${amount} successful. Remaining balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def apply_interest(self):
        self.balance += self.balance * self.interest_rate

class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance, overdraft_limit):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > 0 and amount <= (self.balance + self.overdraft_limit):
            self.balance -= amount
            print(f"Withdrawal of ${amount} from checking account successful. Remaining balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or overdraft limit exceeded.")

class CreditCardAccount(BankAccount):
    def __init__(self, account_number, balance, credit_limit):
        super().__init__(account_number, balance)
        self.credit_limit = credit_limit

    def withdraw(self, amount):
        if amount > 0 and amount <= (self.balance + self.credit_limit):
            self.balance -= amount
            print(f"Withdrawal of ${amount} from credit card successful. Remaining balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or credit limit exceeded.")

# Demonstrate polymorphism
def perform_withdrawal(account, amount):
    account.withdraw(amount)

# Create instances of different account types
savings_account = SavingsAccount(account_number="SA123", balance=1000, interest_rate=0.02)
checking_account = CheckingAccount(account_number="CA456", balance=2000, overdraft_limit=500)
credit_card_account = CreditCardAccount(account_number="CC789", balance=500, credit_limit=1000)

# Call the withdraw() method on objects of different account types
perform_withdrawal(savings_account, 200)      # Calls the withdraw() method of SavingsAccount
perform_withdrawal(checking_account, 2500)    # Calls the withdraw() method of CheckingAccount
perform_withdrawal(credit_card_account, 800)  # Calls the withdraw() method of CreditCardAccount
```

```python
#19. Explain the role of interfaces and abstract classes in polymorphism, drawing comparisons between them.
#ans:
'''In object-oriented programming, both interfaces and abstract classes play crucial roles in achieving polymorphism by providing a
common structure for classes to adhere to. However, they have some differences in terms of their implementation and use. Let's explore
the roles of interfaces and abstract classes in polymorphism and draw comparisons between them:

1. Abstract Classes:
Role in Polymorphism:
===================
An abstract class in OOP serves as a blueprint for other classes but cannot be instantiated on its own. It may contain abstract 
methods (methods without an implementation) that must be implemented by its subclasses.
Abstract classes provide a common interface by defining a set of methods that must exist in any concrete subclass.
Use Cases:
========
Abstract classes are useful when you want to share code among related classes, providing a common base with shared behavior while leaving certain methods 
to be implemented by subclasses.
'''
from abc import ABC, abstractmethod

class AbstractClass(ABC):
    @abstractmethod
    def method_to_implement(self):
        pass
'''Inheritance:
=================
Subclasses extend the abstract class and provide concrete implementations for its abstract methods.'''
class ConcreteClass(AbstractClass):
    def method_to_implement(self):
        print("Concrete implementation of method_to_implement")

instance = ConcreteClass()
instance.method_to_implement()  # Calls the concrete implementation
'''Interfaces:
Role in Polymorphism:

Interfaces provide a way to define a contract that classes must adhere to. An interface specifies a set of methods without providing 
their implementations.
Classes can implement multiple interfaces, allowing them to fulfill the requirements of various contracts.
Use Cases:

Interfaces are useful when you want to define a common set of methods that multiple classes can implement, irrespective of their 
inheritance hierarchy.'''
from abc import ABC, abstractmethod

class Interface1(ABC):
    @abstractmethod
    def method1(self):
        pass

class Interface2(ABC):
    @abstractmethod
    def method2(self):
        pass
'''Implementation:

Classes implement one or more interfaces, providing concrete implementations for the methods specified in those interfaces.'''
class ConcreteClass(Interface1, Interface2):
    def method1(self):
        print("Concrete implementation of method1")

    def method2(self):
        print("Concrete implementation of method2")

instance = ConcreteClass()
instance.method1()  # Calls the concrete implementation of method1
instance.method2()  # Calls the concrete implementation of method2
'''Comparisons:
Abstract Classes:
=================
Can have both abstract and concrete methods.
Allow sharing of code and common behavior.
Support inheritance, promoting code reuse.
Provide a partial implementation.
Interfaces:
=============
Consist only of abstract methods.
Define contracts without specifying any implementation details.
Allow a class to fulfill multiple contracts through interface implementation.
Promote a separation of concerns and a clear contract.
Python Specifics:
=================
In Python, interfaces are not explicitly defined, but the concept is achieved through abstract classes with abstract methods.
Python supports multiple inheritance, allowing a class to inherit from multiple abstract classes or implement multiple interfaces.
Common Ground:
===============
Both abstract classes and interfaces contribute to achieving polymorphism by providing a common structure that multiple classes can 
adhere to, allowing objects of different types to be treated uniformly through a shared interface.
In summary, both abstract classes and interfaces play essential roles in polymorphism, providing mechanisms for defining common 
structures and contracts. The choice between them depends on the design goals and the specific requirements of your application.
Python, being a dynamically-typed language, provides flexibility in using either approach to achieve polymorphism.'''
```
