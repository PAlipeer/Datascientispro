```python
'''Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary.
Implement methods to calculate the yearly bonus and display employee details.'''
#ans:
class Employee:
    def __init__(self, name, emp_id, salary, bonus):
        self.name = name
        self.id = emp_id
        self.salary = salary
        self.bonus = bonus

    def calculate_salary(self, salary, hours_worked):
        overtime = 0
        if hours_worked > 50:
            overtime = hours_worked - 50
        self.salary = self.salary + (overtime * (self.salary / 50))


    def print_employee_details(self):
        print("\nName: ", self.name)
        print("ID: ", self.id)
        print("Salary: ", self.salary)
        print("bonus: ", self.bonus)
        print("----------------------")


employee1 = Employee("ADAMS", "E7876", 50000,2000)
employee2 = Employee("JONES", "E7499", 45000, 2000)
employee3 = Employee("MARTIN", "E7900", 50000, 2000)
employee4 = Employee("SMITH", "E7698", 55000, 2000)

print("Original Employee Details:")
employee1.print_employee_details()
employee2.print_employee_details()
employee3.print_employee_details()
employee4.print_employee_details()


# Now calculate the overtime of the employees who are eligible:
employee2.calculate_salary(45000, 52)
employee4.calculate_salary(45000, 60)

print("Updated Employee Details:")
employee1.print_employee_details()
employee2.print_employee_details()
employee3.print_employee_details()
employee4.print_employee_details()
```

```python
'''Problem 3: Vehicle Rental Create a class representing a vehicle rental system. 
Implement methods to rent a vehicle, return a vehicle, and display available vehicles.'''
#ans:
class Vehicle:
    def __init__(self, make, model, year, available=True):
        self.make = make
        self.model = model
        self.year = year
        self.available = available

    def __str__(self):
        return f"{self.year} {self.make} {self.model}"


class VehicleRentalSystem:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def rent_vehicle(self):
        available_vehicles = [vehicle for vehicle in self.vehicles if vehicle.available]
        if not available_vehicles:
            print("Sorry, no available vehicles for rent.")
            return None
        else:
            print("Available vehicles:")
            for i, vehicle in enumerate(available_vehicles, start=1):
                print(f"{i}. {vehicle}")
            choice = int(input("Enter the number of the vehicle you want to rent: "))
            rented_vehicle = available_vehicles[choice - 1]
            rented_vehicle.available = False
            print(f"You have rented {rented_vehicle}")
            return rented_vehicle

    def return_vehicle(self, returned_vehicle):
        returned_vehicle.available = True
        print(f"Thank you for returning {returned_vehicle}. The vehicle is now available for rent.")

    def display_available_vehicles(self):
        available_vehicles = [vehicle for vehicle in self.vehicles if vehicle.available]
        if not available_vehicles:
            print("No available vehicles at the moment.")
        else:
            print("Available vehicles:")
            for i, vehicle in enumerate(available_vehicles, start=1):
                print(f"{i}. {vehicle}")


# Example usage
if __name__ == "__main__":
    # Creating a few vehicles
    vehicle1 = Vehicle(make="Toyota", model="Camry", year=2020)
    vehicle2 = Vehicle(make="Honda", model="Civic", year=2021)
    vehicle3 = Vehicle(make="Ford", model="Escape", year=2019)

    # Creating a vehicle rental system
    rental_system = VehicleRentalSystem()

    # Adding vehicles to the system
    rental_system.add_vehicle(vehicle1)
    rental_system.add_vehicle(vehicle2)
    rental_system.add_vehicle(vehicle3)

    # Displaying available vehicles
    rental_system.display_available_vehicles()

    # Renting a vehicle
    rented_vehicle = rental_system.rent_vehicle()

    # Displaying available vehicles after renting
    rental_system.display_available_vehicles()

    # Returning the rented vehicle
    if rented_vehicle:
        rental_system.return_vehicle(rented_vehicle)

    # Displaying available vehicles after returning
    rental_system.display_available_vehicles()
```

```python
'''Problem 4: Library Catalog Create classes representing a library and a book. 
Implement methods to add books to the library, borrow books, and display available books.'''
#ans:
class Book:
    def __init__(self, title, author, available=True):
        self.title = title
        self.author = author
        self.available = available

    def __str__(self):
        return f"{self.title} by {self.author}"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' by {book.author} added to the library.")

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and book.available:
                book.available = False
                print(f"You have borrowed '{book.title}' by {book.author}.")
                return book
        print(f"Sorry, '{title}' is not available for borrowing.")
        return None

    def display_available_books(self):
        available_books = [book for book in self.books if book.available]
        if not available_books:
            print("No available books at the moment.")
        else:
            print("Available books in the library:")
            for i, book in enumerate(available_books, start=1):
                print(f"{i}. {book}")

# Example usage
if __name__ == "__main__":
    # Creating a few books
    book1 = Book(title="The Great Gatsby", author="F. Scott Fitzgerald")
    book2 = Book(title="To Kill a Mockingbird", author="Harper Lee")
    book3 = Book(title="1984", author="George Orwell")

    # Creating a library
    library = Library()

    # Adding books to the library
    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)

    # Displaying available books
    library.display_available_books()

    # Borrowing a book
    borrowed_book = library.borrow_book("The Great Gatsby")

    # Displaying available books after borrowing
    library.display_available_books()

    # Borrowing a book that is not available
    not_available_book = library.borrow_book("To Kill a Mockingbird")

    # Returning a borrowed book
    if borrowed_book:
        borrowed_book.available = True
        print(f"Thank you for returning '{borrowed_book.title}'. The book is now available for borrowing.")

    # Displaying available books after returning
    library.display_available_books()

```

```python
'''Problem 5: Product Inventory Create classes representing a product and an inventory system. 
Implement methods to add products to the inventory, update product quantity, and display available products.'''
#ans:
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.name}, Price: ${self.price}, Quantity: {self.quantity}"

class InventorySystem:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"Product '{product.name}' added to the inventory.")

    def update_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.quantity = new_quantity
                print(f"Quantity for product '{product.name}' updated to {new_quantity}.")
                return
        print(f"Product with ID {product_id} not found in the inventory.")

    def display_available_products(self):
        if not self.products:
            print("No products in the inventory.")
        else:
            print("Available products in the inventory:")
            for product in self.products:
                print(product)

# Example usage
if __name__ == "__main__":
    # Creating a few products
    product1 = Product(product_id=1, name="Laptop", price=800, quantity=10)
    product2 = Product(product_id=2, name="Smartphone", price=500, quantity=15)
    product3 = Product(product_id=3, name="Headphones", price=50, quantity=30)

    # Creating an inventory system
    inventory_system = InventorySystem()

    # Adding products to the inventory
    inventory_system.add_product(product1)
    inventory_system.add_product(product2)
    inventory_system.add_product(product3)

    # Displaying available products
    inventory_system.display_available_products()

    # Updating product quantity
    inventory_system.update_quantity(product_id=2, new_quantity=20)

    # Displaying available products after updating quantity
    inventory_system.display_available_products()
```

```python
'''Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height.
Implement methods to calculate the area and perimeter of the shape.'''
#ans:
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

# Example usage
if __name__ == "__main__":
    # Creating a rectangle
    rectangle = Rectangle(length=5, width=3)

    # Calculating and displaying area
    area = rectangle.calculate_area()
    print(f"Area of the rectangle: {area} square units")

    # Calculating and displaying perimeter
    perimeter = rectangle.calculate_perimeter()
    print(f"Perimeter of the rectangle: {perimeter} units")
```

```python
'''Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. 
Implement methods to calculate the average grade and display student details'''
#ans:
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def calculate_average_grade(self):
        if not self.grades:
            return 0  # Avoid division by zero if no grades are available
        return sum(self.grades) / len(self.grades)

    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {', '.join(map(str, self.grades))}")
        average_grade = self.calculate_average_grade()
        print(f"Average Grade: {average_grade:.2f}")

# Example usage
if __name__ == "__main__":
    # Creating a student
    student1 = Student(student_id=1, name="John Doe", grades=[85, 90, 78, 92, 88])

    # Displaying student details
    student1.display_student_details()

    # Calculating and displaying average grade
    average_grade = student1.calculate_average_grade()
    print(f"Overall Average Grade: {average_grade:.2f}")
```

```python
'''Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject.
Implement methods to send an email and display email details.'''
#ans:
class Email:
    def __init__(self, sender, recipient, subject, body):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.body = body
        self.sent = False

    def send_email(self):
        if not self.sent:
            print(f"Email sent from {self.sender} to {self.recipient}")
            self.sent = True
        else:
            print("This email has already been sent.")

    def display_email_details(self):
        print("Email Details:")
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        print(f"Body: {self.body}")
        print("Status: Sent" if self.sent else "Status: Draft")

# Example usage
if __name__ == "__main__":
    # Creating an email
    email1 = Email(sender="sender@example.com", recipient="recipient@example.com", subject="Meeting Agenda", body="Let's discuss the agenda for the upcoming meeting.")

    # Displaying email details
    email1.display_email_details()

    # Sending the email
    email1.send_email()

    # Displaying email details after sending
    email1.display_email_details()
```

```python
'''Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts.
Implement methods to add posts, display posts, and search for posts by keyword'''
#ans:
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, content):
        post = {"content": content, "author": self.username}
        self.posts.append(post)
        print(f"Post added by {self.username}")

    def display_posts(self):
        if not self.posts:
            print("No posts available.")
        else:
            print(f"Posts by {self.username}:")
            for i, post in enumerate(self.posts, start=1):
                print(f"{i}. {post['content']} - by {post['author']}")

    def search_posts_by_keyword(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post['content'].lower()]
        if not matching_posts:
            print(f"No posts containing '{keyword}' found.")
        else:
            print(f"Posts containing '{keyword}':")
            for i, post in enumerate(matching_posts, start=1):
                print(f"{i}. {post['content']} - by {post['author']}")

# Example usage
if __name__ == "__main__":
    # Creating a social media profile
    profile1 = SocialMediaProfile(username="john_doe")

    # Adding posts to the profile
    profile1.add_post(content="Enjoying a sunny day!")
    profile1.add_post(content="Just had a delicious meal.")

    # Displaying all posts
    profile1.display_posts()

    # Searching for posts by keyword
    profile1.search_posts_by_keyword(keyword="sunny")

    # Adding another post
    profile1.add_post(content="New book recommendation: 'The Great Gatsby'")

    # Displaying all posts after adding a new post
    profile1.display_posts()
```

```python
'''Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. 
Implement methods to add tasks, mark tasks as completed, and display pending tasks.'''
#ans:
class ToDoList:
    def __init__(self):
        self.tasks = []
        self.due_dates = []

    def add_task(self, task, due_date=None):
        self.tasks.append({"task": task, "completed": False})
        self.due_dates.append({"task": task, "due_date": due_date})
        print(f"Task '{task}' added to the ToDo list.")

    def mark_as_completed(self, task):
        for t in self.tasks:
            if t["task"] == task:
                t["completed"] = True
                print(f"Task '{task}' marked as completed.")
                return
        print(f"Task '{task}' not found in the ToDo list.")

    def display_pending_tasks(self):
        pending_tasks = [t["task"] for t in self.tasks if not t["completed"]]
        if not pending_tasks:
            print("No pending tasks at the moment.")
        else:
            print("Pending tasks:")
            for i, task in enumerate(pending_tasks, start=1):
                print(f"{i}. {task}")

# Example usage
if __name__ == "__main__":
    # Creating a ToDo list
    todo_list = ToDoList()

    # Adding tasks to the list
    todo_list.add_task(task="Complete project")
    todo_list.add_task(task="Buy groceries", due_date="2022-01-31")
    todo_list.add_task(task="Read a book")

    # Displaying pending tasks
    todo_list.display_pending_tasks()

    # Marking a task as completed
    todo_list.mark_as_completed(task="Complete project")

    # Displaying pending tasks after marking a task as completed
    todo_list.display_pending_tasks()
```
